### TASK#04: Plot basic signal like unit impulse, unit step, unit samples,exponential, unit ramp.

```
import matplotlib.pyplot as plt 
import numpy as np

fig, ax = plt.subplots(2,2, figsize=(15,10))


  
def Unit_Impulse():
  data = []
  stems = []
  
  for i in range(-10,11):
    if(i==0):
      data.append(1)
    else:
      data.append(0)
    
    
  
  # separating the stem parts 
  for i in range(-10,11):
    stems.append(i)
 
  ax[0,0].set_ylabel('x(n)')   # for label at y-axis 
  
  ax[0,0].set_xlabel('n')   # for label at x-axis 
  
  ax[0,0].set_xlim(-10, 10)   # limit of the values at x axis 
  
  ax[0,0].stem(stems, data)   # required plot
  ax[0,0].set_title('Unit impulse signal')
  
  
  
  
  
def Unit_Step():
  data = []
  stems = []
  
  for i in range(-10,11):
    if(i>=0):
      data.append(1)
    else:
      data.append(0)
      
  # separating the stem parts 
  for i in range(-10,11):
    stems.append(i)

  ax[0,1].set_ylabel('x(n)')   # for label at y-axis 
  
  ax[0,1].set_xlabel('n')   # for label at x-axis 
  
  ax[0,1].set_xlim(-10, 10)   # limit of the values at x axis 
  
  ax[0,1].stem(stems, data)   # required plot
  ax[0,1].set_title('Unit step signal')
  
  
  
  
  
def Unit_Ramp():
  data = []
  stems = []
  
  for i in range(-10,11):
    if(i>=0):
      data.append(i)
    else:
      data.append(0)
  
  # separating the stem parts 
  for i in range(-10,11):
    stems.append(i)

  ax[1,0].set_ylabel('x(n)')   # for label at y-axis 
  
  ax[1,0].set_xlabel('n')   # for label at x-axis 
  
  ax[1,0].set_xlim(-10, 10)   # limit of the values at x axis 
  
  ax[1,0].stem(stems, data)   # required plot
  ax[1,0].set_title('Unit ramp signal')
  
  
  
def Exponential():
  data = []
  stems = []
  a = 1.2
  for i in range(-15,10):
      data.append(a**i)
  
  # separating the stem parts 
  for i in range(-15,10):
    stems.append(i)
    
  ax[1,1].set_ylabel('x(n)')   # for label at y-axis 
  
  ax[1,1].set_xlabel('n')   # for label at x-axis 
  
  
  
  ax[1,1].stem(stems, data)   # required plot
  ax[1,1].set_title('Exponential signal')
 


#dsdas
Unit_Impulse()

Unit_Step()

Unit_Ramp()
  
Exponential()



```

### TASK#08A: Implement Moving average filter, Median Filter, Accumulator

```
import numpy as np
sig = [0,3,2,5,-3,4,5,2,0,6,0]
x = np.arange(-3,8)
print("Signal :"+str(sig))


sz = len(x)
y = np.zeros(sz)

for i in range(1,sz-1):
  val = sig[i-1]+sig[i]+sig[i+1]
  val/=3
  y[i] = val

print("Moving average filter : " + str(y))
#print(x)
#print(y)
plt.figure(figsize=(9,17))
plt.subplot(4,1,1)
plt.stem(x,sig)
plt.title("Figure of signal $x_{1}(n)$")


plt.subplot(4,1,2)
plt.stem(x,y)
plt.title("Figure of moving average filter of signal $x_{1}(n)$")

for i in range(1,sz-1):
  val_arr = [sig[i-1],sig[i],sig[i+1]]
  val_arr.sort()
  y[i] = val_arr[1]
  
print("median filter : " + str(y))
  
plt.subplot(4,1,3)
plt.stem(x,y)
plt.title("Figure of median filter of signal $x_{1}(n)$")



for i in range(sz):
  y[i] = 0
  for j in range(i+1):
    y[i]+=sig[j]
    
print("Accumulator : " + str(y))    
plt.subplot(4,1,4)
plt.stem(x,y)
plt.title("Figure of accumulator of signal $x_{1}(n)$")
plt.show()


```
